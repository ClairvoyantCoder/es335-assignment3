```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, Input
from tensorflow.keras.applications import MobileNetV2, EfficientNetB0
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("=" * 80)
print("MNIST CNN EXPERIMENTS WITH PRETRAINED MODELS")
print("=" * 80)

# ============================================================================
# 1. DATA LOADING AND PREPROCESSING
# ============================================================================
print("\n[1] Loading MNIST Dataset...")

# Load MNIST
mnist = keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize to [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Reshape for CNN (add channel dimension)
X_train_cnn = X_train.reshape(-1, 28, 28, 1)
X_test_cnn = X_test.reshape(-1, 28, 28, 1)

print(f"Training set: {X_train_cnn.shape}, Labels: {y_train.shape}")
print(f"Test set: {X_test_cnn.shape}, Labels: {y_test.shape}")

# For pretrained models, we need RGB images (3 channels) and specific input sizes
# MobileNetV2 and EfficientNetB0 require minimum 32x32 images
def preprocess_for_pretrained(X, target_size=32):
    """Resize grayscale MNIST to RGB for pretrained models"""
    # Resize from 28x28 to target_size x target_size
    X_resized = tf.image.resize(X[..., np.newaxis], [target_size, target_size])
    # Convert grayscale to RGB by repeating channel 3 times
    X_rgb = tf.repeat(X_resized, 3, axis=-1)
    return X_rgb.numpy()

print("\n[2] Preprocessing for pretrained models...")
X_train_pretrained = preprocess_for_pretrained(X_train, target_size=32)
X_test_pretrained = preprocess_for_pretrained(X_test, target_size=32)
print(f"Pretrained input shape: {X_train_pretrained.shape}")

# ============================================================================
# 3. BUILD SIMPLE CNN
# ============================================================================
print("\n[3] Building Simple CNN...")
print("Architecture: Conv(32, 3x3) -> MaxPool -> Dense(128) -> Dense(10)")

def create_simple_cnn():
    """
    Simple CNN Architecture:
    - Conv2D: 32 filters, 3x3 kernel, ReLU
    - MaxPooling2D: 2x2
    - Flatten
    - Dense: 128 neurons, ReLU
    - Dense: 10 neurons, Softmax
    """
    model = models.Sequential([
        # Convolutional layer
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1), name='conv1'),

        # Max pooling layer
        layers.MaxPooling2D((2, 2), name='maxpool1'),

        # Flatten for fully connected layers
        layers.Flatten(name='flatten'),

        # Fully connected layer
        layers.Dense(128, activation='relu', name='fc1'),

        # Output layer
        layers.Dense(10, activation='softmax', name='output')
    ], name='SimpleCNN')

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

simple_cnn = create_simple_cnn()
print("\nSimple CNN Architecture:")
simple_cnn.summary()

# ============================================================================
# 4. TRAIN SIMPLE CNN
# ============================================================================
print("\n[4] Training Simple CNN...")
history_cnn = simple_cnn.fit(
    X_train_cnn, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)

print("\n✓ Training completed!")

# ============================================================================
# 5. BUILD MOBILENETV2 MODEL
# ============================================================================
print("\n[5] Building MobileNetV2 Model...")

def create_mobilenet_model():
    """MobileNetV2 pretrained on ImageNet, adapted for MNIST"""
    # Load pretrained MobileNetV2 (without top classification layer)
    base_model = MobileNetV2(
        input_shape=(32, 32, 3),
        include_top=False,
        weights='imagenet',
        pooling='avg'
    )

    # Freeze base model layers (use pretrained features)
    base_model.trainable = False

    # Add custom classification head
    inputs = Input(shape=(32, 32, 3))
    x = base_model(inputs, training=False)
    x = layers.Dense(128, activation='relu')(x)
    outputs = layers.Dense(10, activation='softmax')(x)

    model = models.Model(inputs=inputs, outputs=outputs, name='MobileNetV2')

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

mobilenet_model = create_mobilenet_model()
print("\nMobileNetV2 Architecture:")
mobilenet_model.summary()

# ============================================================================
# 6. TRAIN MOBILENETV2
# ============================================================================
print("\n[6] Training MobileNetV2 (Transfer Learning)...")
history_mobilenet = mobilenet_model.fit(
    X_train_pretrained, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)

print("\n✓ Training completed!")

# ============================================================================
# 7. BUILD EFFICIENTNETB0 MODEL
# ============================================================================
print("\n[7] Building EfficientNetB0 Model...")

def create_efficientnet_model():
    """EfficientNetB0 pretrained on ImageNet, adapted for MNIST"""
    # Load pretrained EfficientNetB0 (without top classification layer)
    base_model = EfficientNetB0(
        input_shape=(32, 32, 3),
        include_top=False,
        weights='imagenet',
        pooling='avg'
    )

    # Freeze base model layers
    base_model.trainable = False

    # Add custom classification head
    inputs = Input(shape=(32, 32, 3))
    x = base_model(inputs, training=False)
    x = layers.Dense(128, activation='relu')(x)
    outputs = layers.Dense(10, activation='softmax')(x)

    model = models.Model(inputs=inputs, outputs=outputs, name='EfficientNetB0')

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

efficientnet_model = create_efficientnet_model()
print("\nEfficientNetB0 Architecture:")
efficientnet_model.summary()

# ============================================================================
# 8. TRAIN EFFICIENTNETB0
# ============================================================================
print("\n[8] Training EfficientNetB0 (Transfer Learning)...")
history_efficientnet = efficientnet_model.fit(
    X_train_pretrained, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)

print("\n✓ Training completed!")

# ============================================================================
# 9. EVALUATION - ACCURACY AND F1-SCORE
# ============================================================================
print("\n[9] Evaluating All Models...")
print("=" * 80)

# Simple CNN
print("\nEvaluating Simple CNN...")
start_time = time.time()
cnn_pred = simple_cnn.predict(X_test_cnn, verbose=0)
cnn_time = time.time() - start_time
cnn_pred_classes = np.argmax(cnn_pred, axis=1)
cnn_accuracy = accuracy_score(y_test, cnn_pred_classes)
cnn_f1 = f1_score(y_test, cnn_pred_classes, average='weighted')
cnn_params = simple_cnn.count_params()

print(f"✓ Accuracy: {cnn_accuracy:.4f}")
print(f"✓ F1-Score: {cnn_f1:.4f}")
print(f"✓ Parameters: {cnn_params:,}")
print(f"✓ Inference Time: {cnn_time:.4f}s ({cnn_time*1000/len(X_test):.2f}ms per image)")

# MobileNetV2
print("\nEvaluating MobileNetV2...")
start_time = time.time()
mobilenet_pred = mobilenet_model.predict(X_test_pretrained, verbose=0)
mobilenet_time = time.time() - start_time
mobilenet_pred_classes = np.argmax(mobilenet_pred, axis=1)
mobilenet_accuracy = accuracy_score(y_test, mobilenet_pred_classes)
mobilenet_f1 = f1_score(y_test, mobilenet_pred_classes, average='weighted')
mobilenet_params = mobilenet_model.count_params()

print(f"✓ Accuracy: {mobilenet_accuracy:.4f}")
print(f"✓ F1-Score: {mobilenet_f1:.4f}")
print(f"✓ Parameters: {mobilenet_params:,}")
print(f"✓ Inference Time: {mobilenet_time:.4f}s ({mobilenet_time*1000/len(X_test):.2f}ms per image)")

# EfficientNetB0
print("\nEvaluating EfficientNetB0...")
start_time = time.time()
efficientnet_pred = efficientnet_model.predict(X_test_pretrained, verbose=0)
efficientnet_time = time.time() - start_time
efficientnet_pred_classes = np.argmax(efficientnet_pred, axis=1)
efficientnet_accuracy = accuracy_score(y_test, efficientnet_pred_classes)
efficientnet_f1 = f1_score(y_test, efficientnet_pred_classes, average='weighted')
efficientnet_params = efficientnet_model.count_params()

print(f"✓ Accuracy: {efficientnet_accuracy:.4f}")
print(f"✓ F1-Score: {efficientnet_f1:.4f}")
print(f"✓ Parameters: {efficientnet_params:,}")
print(f"✓ Inference Time: {efficientnet_time:.4f}s ({efficientnet_time*1000/len(X_test):.2f}ms per image)")

# ============================================================================
# 10. COMPARISON TABLE
# ============================================================================
print("\n" + "=" * 80)
print("COMPREHENSIVE MODEL COMPARISON")
print("=" * 80)
print(f"{'Model':<20} {'Accuracy':<12} {'F1-Score':<12} {'Parameters':<15} {'Time (s)':<12} {'ms/img':<10}")
print("-" * 80)
print(f"{'Simple CNN':<20} {cnn_accuracy:<12.4f} {cnn_f1:<12.4f} {cnn_params:<15,} {cnn_time:<12.4f} {cnn_time*1000/len(X_test):<10.2f}")
print(f"{'MobileNetV2':<20} {mobilenet_accuracy:<12.4f} {mobilenet_f1:<12.4f} {mobilenet_params:<15,} {mobilenet_time:<12.4f} {mobilenet_time*1000/len(X_test):<10.2f}")
print(f"{'EfficientNetB0':<20} {efficientnet_accuracy:<12.4f} {efficientnet_f1:<12.4f} {efficientnet_params:<15,} {efficientnet_time:<12.4f} {efficientnet_time*1000/len(X_test):<10.2f}")
print("=" * 80)

# ============================================================================
# 11. CONFUSION MATRICES
# ============================================================================
print("\n[10] Generating Confusion Matrices...")

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

models_results = [
    ("Simple CNN", cnn_pred_classes, cnn_accuracy),
    ("MobileNetV2", mobilenet_pred_classes, mobilenet_accuracy),
    ("EfficientNetB0", efficientnet_pred_classes, efficientnet_accuracy)
]

for idx, (name, pred, acc) in enumerate(models_results):
    cm = confusion_matrix(y_test, pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx], cbar=False)
    axes[idx].set_title(f'{name}\nAccuracy: {acc:.4f}', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel('Predicted')
    axes[idx].set_ylabel('Actual')

plt.tight_layout()
plt.savefig('cnn_confusion_matrices.png', dpi=300, bbox_inches='tight')
print("✓ Saved: cnn_confusion_matrices.png")
plt.show()

# ============================================================================
# 12. PARAMETER COMPARISON VISUALIZATION
# ============================================================================
print("\n[11] Creating Parameter Comparison Visualization...")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

models = ['Simple CNN', 'MobileNetV2', 'EfficientNetB0']
params = [cnn_params, mobilenet_params, efficientnet_params]
accuracies = [cnn_accuracy, mobilenet_accuracy, efficientnet_accuracy]
times = [cnn_time, mobilenet_time, efficientnet_time]

# Bar chart for parameters
colors = ['#3498db', '#e74c3c', '#2ecc71']
bars1 = axes[0].bar(models, params, color=colors, alpha=0.7, edgecolor='black')
axes[0].set_ylabel('Number of Parameters', fontsize=12, fontweight='bold')
axes[0].set_title('Model Size Comparison', fontsize=14, fontweight='bold')
axes[0].set_yscale('log')
axes[0].grid(axis='y', alpha=0.3)

# Add value labels on bars
for bar, param in zip(bars1, params):
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width()/2., height,
                f'{param:,}',
                ha='center', va='bottom', fontsize=10, fontweight='bold')

# Scatter plot: Accuracy vs Inference Time
# Print values for debugging
print("\nScatter Plot Data:")
for i, (m, t, a) in enumerate(zip(models, times, accuracies)):
    print(f"{m}: Time={t:.4f}s, Accuracy={a:.4f}")

# Plot each point individually with different sizes for visibility
for i, (t, a, color, model) in enumerate(zip(times, accuracies, colors, models)):
    axes[1].scatter(t, a, s=500, c=color, alpha=0.7, edgecolors='black',
                   linewidth=2, zorder=3, label=model)
    # Add labels with smart positioning to avoid overlap
    offset_x = 10 if i == 0 else (-60 if i == 1 else 10)
    offset_y = 10 if i != 1 else -15
    axes[1].annotate(model, (t, a),
                    fontsize=10, fontweight='bold',
                    xytext=(offset_x, offset_y), textcoords='offset points',
                    bbox=dict(boxstyle='round,pad=0.3', facecolor=color, alpha=0.3),
                    arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0', lw=1.5))

axes[1].set_xlabel('Inference Time (seconds)', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Accuracy', fontsize=12, fontweight='bold')
axes[1].set_title('Accuracy vs Speed Trade-off', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3, zorder=1)
axes[1].legend(loc='lower right', fontsize=10)

# Auto-adjust limits to show all points
time_margin = (max(times) - min(times)) * 0.1 if max(times) != min(times) else 0.5
acc_margin = (max(accuracies) - min(accuracies)) * 0.2 if max(accuracies) != min(accuracies) else 0.01
axes[1].set_xlim([min(times) - time_margin, max(times) + time_margin])
axes[1].set_ylim([min(accuracies) - acc_margin, max(accuracies) + acc_margin])

plt.tight_layout()
plt.savefig('cnn_comparison_charts.png', dpi=300, bbox_inches='tight')
print("✓ Saved: cnn_comparison_charts.png")
plt.show()

# ============================================================================
# 13. TRAINING HISTORY VISUALIZATION
# ============================================================================
print("\n[12] Creating Training History Plots...")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

histories = [
    (history_cnn, 'Simple CNN', '#3498db'),
    (history_mobilenet, 'MobileNetV2', '#e74c3c'),
    (history_efficientnet, 'EfficientNetB0', '#2ecc71')
]

for idx, (history, name, color) in enumerate(histories):
    axes[idx].plot(history.history['accuracy'], label='Train', color=color, linewidth=2)
    axes[idx].plot(history.history['val_accuracy'], label='Validation',
                   color=color, linestyle='--', linewidth=2)
    axes[idx].set_xlabel('Epoch', fontsize=11, fontweight='bold')
    axes[idx].set_ylabel('Accuracy', fontsize=11, fontweight='bold')
    axes[idx].set_title(f'{name}\nFinal Val Acc: {history.history["val_accuracy"][-1]:.4f}',
                       fontsize=12, fontweight='bold')
    axes[idx].legend()
    axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('cnn_training_history.png', dpi=300, bbox_inches='tight')
print("✓ Saved: cnn_training_history.png")
plt.show()

# ============================================================================
# 14. DETAILED ANALYSIS
# ============================================================================
print("\n" + "=" * 80)
print("DETAILED ANALYSIS AND OBSERVATIONS")
print("=" * 80)

print("\n1. ACCURACY COMPARISON:")
print(f"   • Simple CNN:      {cnn_accuracy:.4f} - Good baseline performance")
print(f"   • MobileNetV2:     {mobilenet_accuracy:.4f} - Transfer learning benefit")
print(f"   • EfficientNetB0:  {efficientnet_accuracy:.4f} - State-of-the-art architecture")

print("\n2. MODEL SIZE (PARAMETERS):")
print(f"   • Simple CNN:      {cnn_params:,} parameters")
print(f"   • MobileNetV2:     {mobilenet_params:,} parameters ({mobilenet_params/cnn_params:.1f}x larger)")
print(f"   • EfficientNetB0:  {efficientnet_params:,} parameters ({efficientnet_params/cnn_params:.1f}x larger)")
print("   → Simple CNN is most lightweight and deployable")

print("\n3. INFERENCE TIME:")
print(f"   • Simple CNN:      {cnn_time:.4f}s ({cnn_time*1000/len(X_test):.2f}ms/image)")
print(f"   • MobileNetV2:     {mobilenet_time:.4f}s ({mobilenet_time*1000/len(X_test):.2f}ms/image)")
print(f"   • EfficientNetB0:  {efficientnet_time:.4f}s ({efficientnet_time*1000/len(X_test):.2f}ms/image)")
print("   → Simple CNN is fastest for inference")

print("\n4. EFFICIENCY METRICS:")
print(f"   • Simple CNN:      {cnn_accuracy/cnn_params*1000000:.2f} accuracy per million params")
print(f"   • MobileNetV2:     {mobilenet_accuracy/mobilenet_params*1000000:.2f} accuracy per million params")
print(f"   • EfficientNetB0:  {efficientnet_accuracy/efficientnet_params*1000000:.2f} accuracy per million params")

print("\n5. KEY INSIGHTS:")
print("   • Simple CNN: Best for resource-constrained environments")
print("   • MobileNetV2: Good balance of accuracy and efficiency")
print("   • EfficientNetB0: Highest accuracy but more computational cost")
print("   • Transfer Learning: Pretrained models adapt well to MNIST")
print("   • MNIST is relatively simple: Even small CNN achieves ~99% accuracy")

print("\n6. RECOMMENDATIONS:")
print("   • Edge Devices/Mobile: Use Simple CNN (small, fast)")
print("   • Cloud/Server: Use EfficientNetB0 (best accuracy)")
print("   • Balanced Use Case: Use MobileNetV2 (good trade-off)")

print("\n" + "=" * 80)
print("EXPERIMENT COMPLETED SUCCESSFULLY!")
print("=" * 80)
print("\nGenerated files:")
print("  1. cnn_confusion_matrices.png - Performance comparison")
print("  2. cnn_comparison_charts.png - Size and speed trade-offs")
print("  3. cnn_training_history.png - Training progression")
print("\nAll CNN experiments completed! ✓")
```
